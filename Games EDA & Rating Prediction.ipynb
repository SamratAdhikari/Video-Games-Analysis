## Modules

```python
# EDA and analysis
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Model training and prediction
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Miscellaneous
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

## Initialize the dataframe

```python
df = pd.read_csv('assets/games.csv', parse_dates=['released'])
df.sort_values(by=['released'], ascending=True, inplace=True)
df.head()
```

## Data Wrangling

```python
df.columns
```

```python
[df.rename(columns={column: column.capitalize()}, inplace=True) for column in df.columns]    
df.head()
```

```python
df['Released_year'] = df['Released'].dt.year
df['Released_month'] = df['Released'].dt.month
df['Released_day'] = df['Released'].dt.day

df.drop('Released', axis=1, inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

## Exploratory Data Analysis

#### Correlation between the features

```python
corr = df.drop('Name', axis=1).corr()


plt.figure(figsize=(8, 6))
sns.heatmap(data=corr, annot=True, cmap='Greens', cbar=False)
plt.title('Correlation between the Features', fontdict={'size':18, 'family':'Comic Sans MS'})
plt.show()
```

Apparently, the most recent games have more number of reviews, which makes sense.  
Clearly, the features are not related to each other, which is perfect for our ML model.

#### Observing the density distribution of all the significant features

```python
sig_df = df.drop(columns=['Id', 'Name'])
sig_df = sig_df.astype(float)
colors = ['orange', 'purple', 'indianred', 'navy', 'mediumseagreen', 'silver', 'plum', 'yellow']

fig, axs = plt.subplots(4, 2, figsize=(10, 10))

for i, ax in enumerate(axs.flatten()):
    sns.distplot(a=sig_df[sig_df.columns[i]], color=colors[i], ax=ax)
    ax.set_title(sig_df.columns[i])

plt.tight_layout()
plt.show()
```

#### Filling the missing values

```python
np.random.seed(77)
```

```python
df['Released_month'].fillna(value=np.random.randint(1, 13), inplace=True)
df['Released_day'].fillna(value=np.random.randint(1, 31), inplace=True)
df['Released_year'].fillna(value=df['Released_year'].quantile(0.75), inplace=True)

df.isnull().sum()
```

#### Dropping unnecessary columns

```python
df.drop(columns=['Id', 'Name'], inplace=True)
```

## Model Training

```python
X = df.drop('Rating', axis=1)
y = df['Rating']
```

#### Splitting the dataset to train and test sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

#### Choosing the best esitmator

```python
def predict(data, y, alg):
    model = alg().fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    model_name = alg.__name__
    print(model_name)
    
    RMSE = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    print("Model RMSE:", RMSE, "\t\t Model R2 score:", r2)
    print()
```

```python
models = [LinearRegression, Ridge, Lasso, ElasticNet, RandomForestRegressor]

for model in models:
    predict(sig_df, 'Rating', model)
```

`SHEEEEEEEEEEEEESH ! Random Forrest was wayyyyy better in predicting the Rating of Video games !`

#### Comparing the performance between Linear Regression and Random Forest Regressor

```python
model = LinearRegression().fit(X_train, y_train)
lr_preds = model.predict(X_test)

model = RandomForestRegressor().fit(X_train, y_train)
rf_preds = model.predict(X_test)
```

```python
fig = plt.figure(figsize=(10,5))
fig.suptitle("Linear Regression vs Random Forest", fontsize=20)

plt.subplot(1,2,1)
sns.scatterplot(x=y_test, y=lr_preds, color='#272944')
plt.title('Linear Regression predictions Correlation')
plt.xlabel('True labels')
plt.ylabel('Predicted Labels')


plt.subplot(1,2,2)
sns.scatterplot(x=y_test, y=rf_preds, color='#2d7560')
plt.title('Random Forest Regression predictions Correlation')
plt.xlabel('True labels')
plt.ylabel('Predicted Labels')

plt.tight_layout()
plt.show()
```
